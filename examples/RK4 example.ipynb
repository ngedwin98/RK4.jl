```julia
using RK4
```

```julia
function my_ode!(t, x, xdot, params=nothing)
    for kk=2:length(x)
    # some crazy ass ODE
        xdot[kk] = -cos(sum(abs(x))) * x[kk-1]
    end
    xdot[1] = -cos(sum(abs(x))) * x[end]  + sin(t)
end


x0 = zeros(5)
tlist = linspace(0, 100., 1001)

xts = rk4solve(my_ode!, x0, tlist, .01)
```

```julia
using PyPlot
plot(tlist, xts')
```
